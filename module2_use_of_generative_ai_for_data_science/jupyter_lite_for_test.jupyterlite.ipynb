# Test Environment


```python
# Generate statistical description of all features, including 'object' data types
display(df.describe(include='all'))
```

```python
# Create regression plots for numerical attributes against Price
regression_features = ["CPU_frequency", "Screen_Size_inch", "Weight_pounds"]

for feature in regression_features:
    plt.figure(figsize=(10, 6))
    sns.regplot(x=feature, y="Price", data=df)
    plt.title(f'Regression Plot of {feature} vs Price')
    plt.xlabel(feature)
    plt.ylabel('Price')
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.show()
```

```python
# Create box plots for categorical/discrete numerical attributes against Price
boxplot_features = ["Category", "GPU", "OS", "CPU_core", "RAM_GB", "Storage_GB_SSD"]

for feature in boxplot_features:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=feature, y="Price", data=df)
    plt.title(f'Box Plot of {feature} vs Price')
    plt.xlabel(feature)
    plt.ylabel('Price')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()
```

## Authors


```python
from scipy import stats

# Identify numerical columns, excluding 'Price' itself
numerical_cols = df.select_dtypes(include=['number']).columns.tolist()
numerical_cols.remove('Price') # Remove the target variable itself

# Initialize lists to store results
attributes = []
pearson_coefs = []
p_values = []

# Calculate Pearson correlation and p-value for each numerical attribute against 'Price'
for col in numerical_cols:
    pearson_coef, p_value = stats.pearsonr(df[col], df['Price'])
    attributes.append(col)
    pearson_coefs.append(pearson_coef)
    p_values.append(p_value)

# Create a DataFrame to display the results
correlation_df = pd.DataFrame({
    'Attribute': attributes,
    'Pearson_Correlation': pearson_coefs,
    'P_Value': p_values
})

display(correlation_df)
```

```python
# Create a pivot table for GPU, CPU_core, and Price with mean aggregation
pivot_table_df = df.pivot_table(index='GPU', columns='CPU_core', values='Price', aggfunc='mean')

# Display the pivot table
display(pivot_table_df.head())

# Plot a heatmap (pcolor plot) for the pivot table
plt.figure(figsize=(10, 8))
sns.heatmap(pivot_table_df, annot=True, fmt=".0f", cmap="viridis", linewidths=.5)
plt.title('Mean Price by GPU and CPU Core Configuration')
plt.xlabel('CPU Core')
plt.ylabel('GPU')
plt.show()
```
