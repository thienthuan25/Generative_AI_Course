```python
import pandas as pd

# Read the CSV file into a pandas DataFrame, using the first row as headers
df = pd.read_csv(file_name)

# Display the first 5 rows of the DataFrame to verify
display(df.head())
```

### Phát triển và Huấn luyện Mô hình Hồi quy Tuyến tính

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Define source and target variables using the existing variables from the notebook state
source_variable = 'Weight_pounds'
target_variable = 'Price'

# Prepare the data
X = df[[source_variable]] # Independent variable (source_variable)
y = df[target_variable]    # Dependent variable (target_variable)

# Create a Linear Regression model object
lm = LinearRegression()

# Train the model using the training data
lm.fit(X, y)

# Make predictions using the trained model
yhat = lm.predict(X)

display(pd.DataFrame({'Actual': y, 'Predicted': yhat}).head())
```

### Đánh giá Mô hình
Chúng ta sẽ đánh giá mô hình bằng cách sử dụng $R^2$ (R-squared) và Sai số bình phương trung bình (Mean Squared Error - MSE).

```python
# Calculate R-squared
r_squared = r2_score(y, yhat)

# Calculate Mean Squared Error (MSE)
mse = mean_squared_error(y, yhat)

print(f"R-squared: {r_squared}")
print(f"Mean Squared Error: {mse}")

# You can also access the intercept and coefficients
print(f"Intercept: {lm.intercept_}")
print(f"Coefficient for {source_variable}: {lm.coef_[0]}")
```

### Xây dựng Pipeline cho Hồi quy Đa thức với Co giãn Tham số

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, PolynomialFeatures
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd

# Define multiple source variables and one target variable (using the same as previous multi-variable model)
source_variables_pipe = ['CPU_core', 'Screen_Size_inch', 'Weight_pounds']
target_variable_pipe = 'Price'

# Prepare the data
X_pipe = df[source_variables_pipe] # Independent variables
y_pipe = df[target_variable_pipe]    # Dependent variable

# Create the pipeline
# Step 1: Standardize the features (parameter scaling)
# Step 2: Create polynomial features (e.g., degree 2)
# Step 3: Apply Linear Regression
pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('poly', PolynomialFeatures(degree=2)), # You can change the degree here
    ('linear_regression', LinearRegression())
])

# Train the pipeline
pipeline.fit(X_pipe, y_pipe)

# Make predictions using the trained pipeline
yhat_pipe = pipeline.predict(X_pipe)

display(pd.DataFrame({'Actual': y_pipe, 'Predicted (Pipeline)': yhat_pipe}).head())

# Evaluate the pipeline's performance
r_squared_pipe = r2_score(y_pipe, yhat_pipe)
mse_pipe = mean_squared_error(y_pipe, yhat_pipe)

print(f"\n--- Pipeline Model Evaluation (Degree 2) ---")
print(f"R-squared: {r_squared_pipe}")
print(f"Mean Squared Error: {mse_pipe}")
```

## Prepare Data and Create Polynomial Features

### Subtask:
Select the source and target variables from the DataFrame `df`. Then, transform the source variables into polynomial features of degree 2 using `PolynomialFeatures`. This step will create the input data for the Ridge Regression model.


**Reasoning**:
To prepare the data for Ridge Regression, I will import `PolynomialFeatures`, define source and target variables, extract them from the DataFrame, instantiate `PolynomialFeatures` with degree 2, and then transform the source variables.



```python
from sklearn.preprocessing import PolynomialFeatures

# 2. Define a list of source variables and the target variable
source_variables_ridge = ['CPU_core', 'Screen_Size_inch', 'Weight_pounds']
target_variable_ridge = 'Price'

# 3. Create a new DataFrame X_poly_ridge containing only the defined source variables
X_poly_ridge = df[source_variables_ridge]

# 4. Create a new Series y_ridge containing the target variable from df
y_ridge = df[target_variable_ridge]

# 5. Instantiate a PolynomialFeatures object with degree=2
poly_transformer = PolynomialFeatures(degree=2)

# 6. Transform the X_poly_ridge DataFrame into polynomial features
X_poly_transformed = poly_transformer.fit_transform(X_poly_ridge)

print("Polynomial features created successfully.")
print(f"Shape of original source variables (X_poly_ridge): {X_poly_ridge.shape}")
print(f"Shape of transformed polynomial features (X_poly_transformed): {X_poly_transformed.shape}")
```

**Reasoning**:
Now that the polynomial features are created, the next step is to perform a grid search with cross-validation to find the optimal `alpha` hyperparameter for Ridge Regression, as specified in the main task. This involves importing necessary libraries (`Ridge`, `GridSearchCV`), defining the `alpha` parameter grid, and then setting up and executing `GridSearchCV`.



```python
from sklearn.linear_model import Ridge
from sklearn.model_selection import GridSearchCV

# Define the Ridge Regression model
ridge_model = Ridge()

# Define the hyperparameter grid for alpha
param_grid = {'alpha': [0.001, 0.01, 0.1, 1, 10, 100]}

# Set up GridSearchCV with 5-fold cross-validation
grid_search = GridSearchCV(ridge_model, param_grid, cv=5, scoring='r2')

# Fit GridSearchCV to find the best alpha
grid_search.fit(X_poly_transformed, y_ridge)

print("Grid search completed.")
print(f"Best alpha found: {grid_search.best_params_['alpha']}")
print(f"Best R-squared score during grid search: {grid_search.best_score_}")

# Store the best model
best_ridge_model = grid_search.best_estimator_
```

**Reasoning**:
Now that the optimal `alpha` has been found, the next step is to evaluate the performance of the best Ridge Regression model by calculating its R-squared and Mean Squared Error (MSE) on the full dataset.



```python
from sklearn.metrics import mean_squared_error, r2_score

# Make predictions using the best Ridge model
yhat_ridge = best_ridge_model.predict(X_poly_transformed)

# Calculate R-squared for the best Ridge model
r_squared_ridge = r2_score(y_ridge, yhat_ridge)

# Calculate Mean Squared Error (MSE) for the best Ridge model
mse_ridge = mean_squared_error(y_ridge, yhat_ridge)

print(f"--- Best Ridge Regression Model (Polynomial Degree 2) ---")
print(f"Optimal Alpha: {grid_search.best_params_['alpha']}")
print(f"R-squared: {r_squared_ridge}")
print(f"Mean Squared Error: {mse_ridge}")
```

Từ bảng trên, chúng ta có thể thấy:

1.  **Mô hình Hồi quy Tuyến tính Đơn biến** là mô hình kém hiệu quả nhất, với R-squared rất thấp cho thấy 'Weight_pounds' một mình có rất ít khả năng giải thích 'Price' một cách tuyến tính.
2.  **Mô hình Hồi quy Tuyến tính Đa biến** cho thấy sự cải thiện đáng kể về R-squared và giảm MSE khi chúng ta thêm các biến 'CPU_core' và 'Screen_Size_inch'.
3.  Các mô hình **Hồi quy Đa thức** tiếp tục cải thiện hiệu suất. R-squared tăng lên và MSE giảm khi bậc của đa thức tăng từ 2 lên 5.
4.  **Mô hình Hồi quy Đa thức Bậc 5** là mô hình có hiệu suất tốt nhất trong số các mô hình hồi quy đa thức thuần túy, với R-squared cao nhất (0.3023) và MSE thấp nhất (229408.80) trong nhóm này. Điều này cho thấy mối quan hệ giữa các biến nguồn và Price có lẽ là phi tuyến tính.
5.  **Mô hình Ridge Hồi quy với Đa thức Bậc 2** (tối ưu hóa `alpha` = 1) đã cho thấy một bước nhảy vọt đáng kể về hiệu suất, đạt R-squared là 0.4635 và MSE là 176385.64. Điều này cho thấy việc kết hợp các tính năng đa thức với chuẩn hóa L2 của Ridge Regression đã giúp cải thiện đáng kể khả năng tổng quát hóa và độ chính xác của mô hình so với các mô hình tuyến tính và đa thức đơn giản trước đó.

Tuy nhiên, cần lưu ý rằng mặc dù mô hình Ridge Regression là tốt nhất trong số các mô hình đã thử nghiệm, R-squared vẫn chỉ khoảng 0.46. Điều này ngụ ý rằng các biến hiện tại, ngay cả khi được biến đổi đa thức và với chuẩn hóa Ridge, chỉ giải thích được khoảng 46% sự biến động của giá, và có thể có nhiều yếu tố khác ảnh hưởng đến giá laptop chưa được đưa vào mô hình hoặc có thể cần các kỹ thuật mô hình hóa phức tạp hơn.
